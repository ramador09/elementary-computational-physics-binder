## Theory in brief

### Why "paradoxes" are not contradictions

Every special-relativity paradox is a clash between Newtonian intuition and the correct
relativistic account, and every one resolves with a tool already in hand,

```{math}
:label: eq-no-paradox
\text{resolution} = \text{proper time along the worldline (4.3)} \ \text{ or } \ \text{relativity of simultaneity (4.1)} .
```

We settle each by computation, not rhetoric.

### Proper time along a worldline

The time a clock reads is the **proper time**, the "length" of its worldline in the
spacetime metric,

```{math}
:label: eq-proper-time
\tau=\int d\tau=\int\sqrt{1-\frac{v(t)^2}{c^2}}\,dt=\int\frac{dt}{\gamma(t)} .
```

Because of the minus sign in the metric, the **inertial (straight) worldline between two
events has the maximum proper time**, the exact reverse of the Euclidean rule that a
straight line is the shortest path. This single fact resolves the twin paradox.

### The twin paradox

One twin travels out and back at high speed while the other stays home,

```{math}
:label: eq-twin
\tau_{\rm travel}=\int\sqrt{1-\beta(t)^2}\,dt < \tau_{\rm home}=T .
```

The apparent symmetry â€” "each sees the other's clock run slow" â€” is false, because the
worldlines are not symmetric: the traveller *accelerates* to turn around, and the
stay-home twin does not. Integrating proper time along each worldline settles it with no
ambiguity. We do both the idealized instant turnaround and an honest smooth acceleration.

### The pole-barn (ladder) paradox

A pole of proper length $L_p$ longer than a barn of proper length $L_b$ "fits" inside when
it runs through fast, length-contracted, in the barn frame; but in the pole's frame the
*barn* is contracted and the pole cannot fit,

```{math}
:label: eq-pole-barn
\text{barn frame: } L_p/\gamma < L_b, \qquad \text{pole frame: } L_p > L_b/\gamma .
```

The resolution is that "the pole is fully inside with both doors shut simultaneously" is a
statement about simultaneity. The two door-closing events are simultaneous in the barn
frame but separated by $\Delta t'=\gamma vL_b/c^2$ in the pole frame, so the two frames
disagree about "at once" without disagreeing about anything physical.

### The relativistic Doppler effect

Not a paradox but a real, measurable consequence, combining the classical Doppler geometry
with time dilation,

```{math}
:label: eq-doppler
\frac{f_{\rm obs}}{f_{\rm src}}=\sqrt{\frac{1\pm\beta}{1\mp\beta}} \ \text{(approach/recession)}, \qquad \frac{f_{\rm obs}}{f_{\rm src}}=\frac{1}{\gamma} \ \text{(transverse)} .
```

The transverse shift is a pure redshift even at $90^\circ$, from time dilation alone, with
no classical analog â€” the Ivesâ€“Stilwell confirmation of time dilation.

````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from scipy.integrate import quad

from ecp import draw, validate
from ecp.animate import show

MU0 = 4.0e-7 * np.pi  # vacuum permeability, TÂ·m/A
EPS0 = 8.8541878128e-12  # vacuum permittivity, F/m
C_LIGHT = 1.0 / np.sqrt(MU0 * EPS0)  # speed of light, m/s
ACCENT, INK, SOFT = draw.ACCENT, draw.INK, draw.SOFT


def gamma(beta):
    """The Lorentz factor Î³ = 1/sqrt(1 âˆ’ Î²^2) as a function of Î² = v/c (eq-proper-time).

    Parameters
    ----------
    beta : float or numpy.ndarray
        Speed as a fraction of c (|Î²| < 1).

    Returns
    -------
    float or numpy.ndarray
        The Lorentz factor, matching the shape of ``beta``.
    """
    return 1.0 / np.sqrt(1.0 - np.asarray(beta) ** 2)


def proper_time(beta_of_t, t0, t1, n=2001):
    """The proper time along a worldline: the integral of sqrt(1 âˆ’ Î²(t)^2) dt (eq-proper-time).

    Evaluates the elapsed clock time for a worldline with instantaneous speed
    Î²(t) = v(t)/c, by the trapezoidal rule (``np.trapezoid``) over a dense
    grid of ``n`` samples. Direction does not matter, only speed, since the
    integrand depends on Î²^2.

    Parameters
    ----------
    beta_of_t : callable
        Function returning the speed Î²(t) (a fraction of c) at time ``t``.
    t0, t1 : float
        Start and end coordinate times (any consistent unit; the result shares it).
    n : int, optional
        Number of grid samples (default ``2001``).

    Returns
    -------
    float
        The proper time elapsed between ``t0`` and ``t1``.
    """
    t = np.linspace(t0, t1, n)
    integrand = np.sqrt(1.0 - beta_of_t(t) ** 2)
    return float(np.trapezoid(integrand, t))
````

## Exercise 7 â€” The transverse Doppler effect

Here is the cleanest signature of time dilation in all of optics. Classically, a source
moving *across* the line of sight â€” at the instant it is exactly abeam, $90^\circ$ â€” produces

classical analog whatsoever {eq}`eq-doppler`. The moving source's clock simply runs slow, so

Stilwell in 1938, is a direct experimental confirmation of time dilation.
